```python
import pandas as pd

df = pd.read_table(
    'data/water_solublity.tsv', sep='\t', header=None
).rename(
    columns={
        0: 'SMILES',
        2: 'logS'
    }
).drop([1],axis=1)

smiles = list(df.SMILES)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors.MoleculeDescriptors import MolecularDescriptorCalculator

# 計算対象とする記述子
descriptors = [descriptor_information[0] for descriptor_information in Descriptors.descList]
descriptor_calculator = MolecularDescriptorCalculator(descriptors)
```

```python

features = []
for smile in smiles:
    try:
        molecule = Chem.MolFromSmiles(smile)
        feature = descriptor_calculator.CalcDescriptors(molecule)
    except:
        print(smile)
 
#features = [descriptor_calculator.CalcDescriptors(Chem.MolFromSmiles(smile)) for smile in smiles]

#feature_df = pd.DataFrame(features, index=dataset.index, columns=descriptor_names)
```
